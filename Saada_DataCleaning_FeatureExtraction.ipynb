Here’s a detailed explanation of each column from your dataset:

### 1. **(Parent) ASIN:**
   - **Definition:** The Amazon Standard Identification Number (ASIN) for the parent product. Parent ASINs group related product variations (e.g., size, color).
   - **Usage:** Helps in identifying and grouping similar products in the dataset.

### 2. **(Child) ASIN:**
   - **Definition:** The ASIN for a specific variation (child) of a product. Each variant, such as different sizes or colors, has its own child ASIN.
   - **Usage:** Useful for analyzing specific product variations separately.

### 3. **Title:**
   - **Definition:** The name or description of the product.
   - **Usage:** Provides product identification for easier interpretation.

### 4. **SKU (Stock Keeping Unit):**
   - **Definition:** A unique identifier for products in inventory.
   - **Usage:** Useful for inventory management and tracking specific items.

### 5. **Sessions - Total:**
   - **Definition:** The total number of visits (sessions) to the product detail page.
   - **Usage:** Indicates the level of traffic a product is receiving from potential customers.

### 6. **Sessions - Total - B2B:**
   - **Definition:** The total number of sessions from Business-to-Business (B2B) customers.
   - **Usage:** Helps in distinguishing B2B engagement from total customer sessions.

### 7. **Session Percentage - Total:**
   - **Definition:** The percentage of total sessions for the product in relation to all sessions in the store.
   - **Usage:** Measures how much attention the product is getting compared to other products in the store.

### 8. **Session Percentage - Total - B2B:**
   - **Definition:** The percentage of total B2B sessions in relation to all B2B sessions in the store.
   - **Usage:** Measures the product's visibility and interest among B2B customers.

### 9. **Page Views - Total:**
   - **Definition:** The total number of page views for the product detail page.
   - **Usage:** Indicates the interest level of the product by tracking how often customers view its details.

### 10. **Page Views - Total - B2B:**
   - **Definition:** The number of page views from B2B customers.
   - **Usage:** Helps gauge the interest of B2B customers in the product.

### 11. **Page Views Percentage - Total:**
   - **Definition:** The percentage of page views for the product compared to all other products in the store.
   - **Usage:** Shows how popular the product is in terms of views across the entire product lineup.

### 12. **Page Views Percentage - Total - B2B:**
   - **Definition:** The percentage of page views from B2B customers compared to all B2B page views.
   - **Usage:** Measures B2B customer interest relative to other products in the B2B segment.

### 13. **Featured Offer (Buy Box) Percentage:**
   - **Definition:** The percentage of times the product was the featured "Buy Box" offer.
   - **Usage:** Being in the Buy Box typically increases the chances of a sale, so a higher percentage is desirable.

### 14. **Featured Offer (Buy Box) Percentage - B2B:**
   - **Definition:** The percentage of times the product was the Buy Box offer specifically for B2B customers.
   - **Usage:** Similar to the overall Buy Box percentage but focused on B2B customers.

### 15. **Units Ordered:**
   - **Definition:** The total number of units of the product that were ordered.
   - **Usage:** A key metric to gauge product sales performance.

### 16. **Units Ordered - B2B:**
   - **Definition:** The total number of units ordered by B2B customers.
   - **Usage:** Helps identify how much of the product is being purchased by businesses.

### 17. **Unit Session Percentage:**
   - **Definition:** The percentage of sessions that resulted in an order (i.e., conversion rate).
   - **Usage:** A critical metric for understanding how effective product pages are at converting visits into purchases.

### 18. **Unit Session Percentage - B2B:**
   - **Definition:** The percentage of B2B sessions that resulted in an order.
   - **Usage:** Helps evaluate how well the product page converts B2B visits into sales.

### 19. **Ordered Product Sales:**
   - **Definition:** The total revenue generated from ordered units of the product.
   - **Usage:** A key financial metric to assess product performance in terms of sales.

### 20. **Ordered Product Sales - B2B:**
   - **Definition:** The total revenue generated from B2B customers.
   - **Usage:** Shows the financial contribution of B2B sales to overall product revenue.

### 21. **Total Order Items:**
   - **Definition:** The total number of items ordered (includes multiple quantities in single orders).
   - **Usage:** Measures the total number of individual items sold.

### 22. **Total Order Items - B2B:**
   - **Definition:** The total number of items ordered by B2B customers.
   - **Usage:** Provides insight into the volume of B2B sales.

```python
df.loc[268, 'SKU'] = df.loc[268, 'SKU'].replace(' ', '')
```

```python
def extract_color(sku):
    if '_' in sku:
        parts = sku.split('_')
        if len(parts) == 3:
            # For SKUs like SD_CPWH_XL
            color = parts[1][-2:]  # The first two characters of the last part
        elif len(parts) == 2:
            # For SKUs like SDRSTMA_3XL
            color = parts[0][-2:]  # The last two characters of the first part (before the underscore)
    else:
        # Handle cases without underscores if needed
        color = None

    return color

# Apply the function to the DataFrame
df['Color Code'] = df['SKU'].apply(extract_color)

df
```

```python
color_mapping = {
    'BL': 'Black',
    'WH': 'White',
    'MA': 'Maroon',
    'BE': 'Beige',
    'OG': 'Orange',
    'RT': 'Rust',
    'LI': 'Light Blue',
    'NB': 'Navy Blue',
    'EC': 'Electric Blue',
    'EB': 'Emerald Blue',
    'CG': 'Charcoal Grey',
    'GR': 'Grey',
    'YE': 'Yellow',
    'SP': 'Steel Pink',
    'TU': 'Turquoise',
    'SG': 'Sea Green',
    'PE': 'Peach',
    'JA': 'Jade',
    'MG': 'Magenta',
    'RO': 'Rose',
    'HP': 'Hot Pink',
    'BR': 'Brown',
    'BG': 'Beige Gold',
    'WI': 'Wine',
    'PB': 'Powder Blue',
    'CB': 'Cobalt Blue',
    'TE': 'Teal',
    'RB': 'Royal Blue',
    'SB': 'Sky Blue',
    'AS': 'Ash',
    'MI': 'Mint'
}

# Map color codes to full names
df['Color'] = df['Color Code'].map(color_mapping)

# Display the updated DataFrame
df
```
